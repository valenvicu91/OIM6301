```python
# ---------- Exercise 01.1: Fermat's Last Theorem ----------

def check_fermat(a, b, c, n):
    """Prints whether a^n + b^n == c^n for n > 2 (Fermat’s check)."""
    if n > 2 and a**n + b**n == c**n:
        print("Holy smokes, Fermat was wrong!")
    else:
        print("No, that doesn't work.")

def test_fermat():
    """Prompts the user for a, b, c, n and checks Fermat's theorem."""
    a = int(input("Enter value for a: "))
    b = int(input("Enter value for b: "))
    c = int(input("Enter value for c: "))
    n = int(input("Enter value for n: "))
    check_fermat(a, b, c, n)

# ---------- Exercise 01.2: BMI Calculator ----------

def calculate_bmi(weight, height):
    """Return BMI given weight (kg) and height (m)."""
    return weight / (height ** 2)

def get_bmi_category():
    """
    Prompt for weight (kg) and height (m), print BMI, and return the category.
      <18.5 Underweight, 18.5–<25 Normal, 25–<30 Overweight, >=30 Obesity
    """
    weight = float(input("Enter your weight in kilograms: "))
    height = float(input("Enter your height in meters: "))
    bmi = calculate_bmi(weight, height)
    print(f"Your BMI is {bmi:.2f}")

    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obesity"
```

```python
# ---------- Exercise 02.1: Factorial (recursive) ----------

def factorial(n):
    """Return n! for integer n >= 0."""
    if n == 0:
        return 1
    return n * factorial(n - 1)

# Examples:
# print(factorial(5))   # 120
# print(factorial(0))   # 1


# ----------Exercise 02.2: Fibonacci (recursive) ----------
# 0-indexed: F0 = 0, F1 = 1

def fibonacci(n):
    """Return the nth Fibonacci number."""
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

# Examples:
# print(fibonacci(6))   # 8
# print(fibonacci(10))  # 55


# ---------- Exercise 02.3: GCD using Euclid’s algorithm (recursive) ----------

def gcd(a, b):
    """Return the greatest common divisor of positive integers a and b."""
    if b == 0:
        return a
    return gcd(b, a % b)

# Examples:
# print(gcd(2, 12))   # 2
# print(gcd(6, 12))   # 6
# print(gcd(9, 12))   # 3
# print(gcd(17, 12))  # 1


# ----------Exercise  02.4 (Optional): Tower of Hanoi (recursive) ----------

def move(n, source, bridge, destination):
    """
    Print steps to move n disks from 'source' to 'destination'
    using 'bridge' as auxiliary.
    """
    if n == 1:
        print(f"{source} --> {destination}")
        return
    move(n - 1, source, destination, bridge)
    print(f"{source} --> {destination}")
    move(n - 1, bridge, source, destination)

# Example (uncomment to see expected sequence):
# move(3, 'A', 'B', 'C')
```
