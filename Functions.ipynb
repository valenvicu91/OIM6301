```python
# exercise 05
def my_abs(x):
    # First check the type of argument
    if not isinstance(x, (int, float)):
        raise TypeError("The argument must be an integer or float")
    if x < 0:
        return -x
    else:
        return x

# Test cases
print(my_abs(-10))
print(my_abs(4.5))
print(my_abs("hello"))
```

```python
# Exercise 06
import math

def quadratic(a, b, c):
    # Calculate the discriminant
    discriminant = b**2 - 4*a*c

    if discriminant < 0:
        return None  # No real roots

    # Calculate the two roots
    x1 = (-b + math.sqrt(discriminant)) / (2*a)
    x2 = (-b - math.sqrt(discriminant)) / (2*a)

    return x1, x2

# Test cases
print(quadratic(1, -3, 2))   # roots: (2.0, 1.0)
print(quadratic(1, 2, 1))    # roots: (-1.0, -1.0)
```

```python
# Excercise 07
import math

# Define a function that moves a point (x, y)
# by a given number of steps at a certain angle.

def move(x, y, step, angle=0):
    nx = x + step * math.cos(angle)   # new x-coordinate
    ny = y - step * math.sin(angle)   # new y-coordinate
    return nx, ny                     # returns two values

# Example call
x, y = move(100, 100, 60, math.pi / 6)
print(x, y)
print(f"New position: ({x:.2f}, {y:.2f})")
```
