```python
# Exercise 02

# 1. Seconds in 42 minutes 42 seconds
seconds = 42 * 60 + 42
print("Total seconds:", seconds)

# 2. Miles in 10 kilometers (1 mile = 1.61 km)
kilometers = 10
miles = kilometers / 1.61
print("Miles:", miles)

# 3. Pace (time per mile) and average speed
pace_seconds_per_mile = seconds / miles
pace_min = int(pace_seconds_per_mile // 60)
pace_sec = int(pace_seconds_per_mile % 60)
speed_mph = miles / (seconds / 3600)

print(f"Pace: {pace_min} min {pace_sec} sec per mile")
print(f"Average speed: {speed_mph:.2f} mph")
```
