```python
a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))

print(f"{a} + {b} = {a + b:5}")
print(f"{a} - {b} = {a - b:5}")
print(f"{a} * {b} = {a * b:5}")
print(f"{a} / {b} = {a / b:5.2f}")
print(f"{a} ^ {b} = {a ** b:5}")
```
