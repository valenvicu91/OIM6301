```python

# Volume of a sphere with radius r
# Formula: (4/3) * π * r³
import math

r = 5
volume = (4 / 3) * math.pi * r**3
print(f"1. Volume of a sphere with radius {r} = {volume:.2f} cubic units")

# Total wholesale cost for 60 copies
# Cover price = $24.95
# Bookstores get 40% discount → pay 60% of cover price
# Shipping: $3 for first copy + $0.75 for each additional

cover_price = 24.95
discount = 0.4
wholesale_price = cover_price * (1 - discount)
num_copies = 60
shipping_cost = 3 + (num_copies - 1) * 0.75
total_cost = wholesale_price * num_copies + shipping_cost

print(f"2. Total wholesale cost for {num_copies} copies = ${total_cost:.2f}")

# Water bottle percentage remaining
# Bottle capacity: 750 ml, drank 250 ml → remaining 500 ml
# Percentage remaining = (remaining / total) * 100

total_ml = 750
drank_ml = 250
remaining_ml = total_ml - drank_ml
percent_remaining = (remaining_ml / total_ml) * 100

print(f"3. Percentage of bottle remaining = {percent_remaining:.1f}%")

# Laptop discount percentage
# Original price: $899 → Sale price: $749
# Discount % = ((original - sale) / original) * 100

original_price = 899
sale_price = 749
discount_percent = ((original_price - sale_price) / original_price) * 100

print(f"4. Percentage discount = {discount_percent:.1f}%")
```
